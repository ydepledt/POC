```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd

from shapely.geometry import Point
```

```python
data = pd.read_csv('Crime_Data_from_2020_to_Present.csv')
```

```python
data
```

```python
# Remove rows with missing lat/long values
data = data.dropna(subset=['LAT', 'LON'])

# Remove rows with strange lat/long values
data = data[(data['LAT'] > 33) & (data['LAT'] < 35)]
data = data[(data['LON'] > -119) & (data['LON'] < -117)]
```

```python
data
```

```python
data_reduced = data.head(1000)


# Create a GeoDataFrame
geometry = [Point(xy) for xy in zip(data_reduced['LON'], data_reduced['LAT'])]
gdf = gpd.GeoDataFrame(data_reduced, geometry=geometry)

# Load the map of the USA (update with your own shapefile path)
la_map = gpd.read_file('./LosAngeles/tl_2020_06_tabblock20_LACounty.shp')


# Plotting the map
fig, ax = plt.subplots(figsize=(400, 400))
la_map.boundary.plot(ax=ax, linewidth=1)
gdf.plot(ax=ax, color='red', markersize=50, alpha=0.6)

# Add labels for crime locations
# for x, y in zip(gdf.geometry.x, gdf.geometry.y):
#     ax.text(x, y, fontsize=12, ha='right')

# Set titles and labels
plt.title('Crime Locations in the USA', fontsize=15)
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.savefig('crime_locations.png')
plt.show()
```
